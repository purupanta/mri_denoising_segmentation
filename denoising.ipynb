**The following method is created to plot the original image and the filtered image side by side so that we can compare the result.**

```python
def plot_comparison(img_original, img_filtered, filtered_img_name):

    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 6), sharex=True, sharey=True)
    ax1.imshow(img_original, cmap=plt.gray())
    ax1.set_title('==Original==')
    ax1.axis('off')
    ax2.imshow(img_filtered, cmap=plt.gray())
    ax2.set_title(filtered_img_name)
    ax2.axis('off')
```

**Denoising #1:**<br />
Mean denoising with "cv2" library seems to be producing great denoising result for this image becase the noise in this image seems to be more like salt-pepper

```python
import cv2
# from matplotlib import pyplot as plt

image_bw = cv2.imread('./drive/MyDrive/Colab Notebooks/BMI734/hw1/embryos.jpg',0)
image_colored = cv2.imread('./drive/MyDrive/Colab Notebooks/BMI734/hw1/embryos.jpg',1)

median = cv2.medianBlur(image_colored, 5)
compare = np.concatenate((image_colored, median), axis=1) #side by side comparison
```

```python
# Removing the noise with Mean Denoising

# Denoising as if this image is black-white
noiseless_image_bw = cv2.fastNlMeansDenoising(image_bw, None, 20, 7, 21)
# Plotting the original image and the denoised image
plot_comparison(image_bw, noiseless_image_bw, '==Denoised==')

# Denoising as if this image is colored
noiseless_image_colored = cv2.fastNlMeansDenoisingColored(image_colored,None,20,20,7,21)
# Plotting the original image and the denoised image
plot_comparison(image_colored, noiseless_image_colored, '==Denoised==')








```

**Denoising #2:**<br />
Using Skimage restoration denoise chambolle. I did "not" see it giving good result. It seems to be slightly removing a bit of salt and pepper from the image however.

```python
from skimage.restoration import denoise_tv_chambolle

noisy_image = plt.imread('./drive/MyDrive/Colab Notebooks/BMI734/hw1/embryos.jpg')

# For denoising, apply the total variation filter denoising
denoised_image = denoise_tv_chambolle(noisy_image, channel_axis=10)

# Plot the original noisy image and the denoised image
plot_comparison(noisy_image, denoised_image, '==Denoised==')
```
